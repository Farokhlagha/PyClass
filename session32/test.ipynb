```python
import cv2
import matplotlib.pyplot as plt
```

```python
image_low_contrast = cv2.imread('input/low-cont.png', cv2.IMREAD_GRAYSCALE)
hist = cv2.calcHist([image_low_contrast], [0], None, [256], [0,256])

plt.plot(hist)
plt.show()
```

```python
image_high_contrast = cv2.equalizeHist(image_low_contrast)

hist = cv2.calcHist([image_high_contrast], [0], None, [256], [0,256])
plt.plot(hist)
plt.show()

cv2.imwrite('output/test1.jpg', image_high_contrast)
```
