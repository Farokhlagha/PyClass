Analysis billionaires of world with python

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Billionaire.csv')
data.head()
```

Preprocess

```python

data.isnull().sum()
```

```python
data = data.dropna().reset_index()
```

Process

```python
data = data.dropna().reset_index()
data['NetWorth'] = data['NetWorth'].str.strip('$')
data['NetWorth'] = data['NetWorth'].str.strip('B').to_numpy(dtype=float)
df = data.sort_values(by=['NetWorth'], ascending=False).head(10)
df.head()
```

```python
plt.figure(figsize=(20,6))
plt.bar(df['Name'], df['NetWorth'])
plt.show()
```

```python
df = data['Source'].value_counts().head(10).reset_index()
df
```

```python
plt.pie(df['count'], labels=df['Source'])
plt.title('Top 10 domains become a billionaires')
plt.show()
```

```python
df= data['Industry'].value_counts().reset_index()
df
```

```python
plt.figure(figsize=(24, 8))
plt.pie(df['count'], labels=df['Industry'])
plt.title('Top 10 Industries become a billionaires')
plt.show()
```

```python
df= data['Country'].value_counts().head(10).reset_index()
df
```

```python
plt.pie(df['count'], labels=df['Country'])
plt.title('Top 5 Countries become a billionaires')
plt.show()
```
