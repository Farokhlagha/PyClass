```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('cal_cities.csv')
data.head()
```

```python
plt.scatter(data['Latitude'], 
            data['Longitude'], 
            s=data['area-km2'], 
            alpha=0.5, 
            linewidths=0, 
            c= np.log10(data['pop_april_2010']), 
            cmap='gist_ncar')
plt.colorbar(label= 'log 10 population')
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.title('Area and Population of California Cities')
plt.show()
```
