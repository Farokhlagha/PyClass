```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('input/wolf.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(image, cmap='gray')  # cv2.imshow() in notebook and colab do not work instate apply plt.imshow()
plt.show()   # remove all extra thing on pic
```

```python
_, image_thresh = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)
plt.imshow(image_thresh, cmap='gray')
```

```python
contours, _ = cv2.findContours(image_thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

result = image.copy()
contour = 0
for contour in contours:
    if cv2.contourArea(contour) > 1000:
        x, y, w, h = cv2.boundingRect(contour)
        cv2.rectangle(image, (x, y), (x+w, y+h), (255, 255, 255), 3)

plt.imshow(image, cmap='gray')
```

```python
image_puma = cv2.imread('input/puma.png', cv2.IMREAD_GRAYSCALE)
plt.imshow(image_puma, cmap='gray')  
```

```python
image_puma = 255 - image_puma  # color of background and foreground are changed

plt.imshow(image_puma, cmap='gray')
```

```python
_, image_puma_threshold = cv2.threshold(image_puma, 115, 255, cv2.THRESH_BINARY)
```

```python
contour, _ = cv2.findContours(image_puma, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

result = image_puma.copy()

for contour in contours:
    if cv2.contourArea(contour) >1000:
        x, y, w, h = cv2.boundingRect(contour)
        cv2.rectangle(result, (x,y), (x+w, y+h), (127,127, 127), 3)

plt.imshow(result, cmap='gray')  
```

```python
image_coins = cv2.imread('input/coins.jpg', cv2.IMREAD_GRAYSCALE)
plt.axis('off')  # no x and y axis
plt.imshow(image_coins, cmap='gray')
plt.show()
```

```python
_, image_coins_thresh = cv2.threshold(image_coins, 90, 255, cv2.THRESH_BINARY)
plt.axis('off')
plt.imshow(image_coins_thresh, cmap='gray')
plt.show()
```

```python
contours, hierarchy = cv2.findContours(image_coins_thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
result = image_coins.copy()

for contour in contours:
    x,y,w,h = cv2.boundingRect(contour)
    cv2.rectangle(result, (x,y), (x+w, y+h), (127, 127, 127), 3)

    plt.axis('off')
    plt.imshow(result, cmap='gray')
    plt.show()
```
