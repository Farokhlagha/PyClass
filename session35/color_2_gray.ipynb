```python
import cv2
import numpy as np
import time
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('input/mandril.jpeg')
image.shape
```

```python
# one solution:
start_time = time.time()

R, G, B = cv2.split(image)
R = R.astype(float) # or you can do  R[i,j]//3 + G[i,j]//3 + B[i,j]//3 instead of make float
G = G.astype(float)
B = B.astype(float)

result = np.zeros(image.shape[:2], dtype=np.uint8)

for i in range(image.shape[0]):
    for j in range (image.shape[1]):
        result[i,j]= (R[i,j] + G[i,j] + B[i,j]) // 3

end_time = time.time()
print(end_time - start_time)

plt.imshow(result, cmap='gray')
```

```python
# fast solution
start_time = time.time()

result = np.mean(image, axis=2)  # for dimention 3 is 2 (0,1,2)

end_time = time.time()
print(end_time - start_time)

plt.imshow(result, cmap='gray')
plt.show()
```
